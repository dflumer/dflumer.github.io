```python
import random

def selection_sort(arr):
    # Sort an array using selection sorting
    n = len(arr)
    index_vector = list(range(1, n + 1))  # Initialize index vector

    for i in range(n):
        min_idx = i  # Assume the current position is the minimum

        # Find the actual minimum element in the unsorted part
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j

        # Swap the minimum element with the first element
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
        
        # Do the same swap in the index vector
        index_vector[i], index_vector[min_idx] = index_vector[min_idx], index_vector[i]

    return arr, index_vector

# Generate 25 random real numbers between 0 and 100
random_numbers = [random.randint(0, 100) for x in range(25)]
print("Original Array:", random_numbers)

# Preserved sorted list
save_sort = random_numbers[:]

# Apply the selection sort algorithm
sorted_array, index_vector = selection_sort(save_sort)
print("Sorted Array:", sorted_array)
print("Index Vector:", index_vector)

# Validate with Python's sort
py_sort = sorted(random_numbers)
py_iv = [random_numbers.index(x) + 1 for x in py_sort]
print("\nSorted Array Validation:", py_sort)
print("Index Vector Validation:", py_iv)
```

```python
import random

def bubble_sort(arr):
    """
    Sorts an array using the bubble sort algorithm and returns the sorted array
    and the index vector.
    """
    n = len(arr)
    index_vector = list(range(1, n + 1))  # Initialize index vector

    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                # Swap if the element found is greater than the next element
                arr[j], arr[j+1] = arr[j+1], arr[j]
                index_vector[j], index_vector[j+1] = index_vector[j+1], index_vector[j]

    return arr, index_vector

# Generate 25 random whole numbers between 0 and 100
random_numbers = [random.randint(0, 100) for x in range(25)]
print("Original Array:", random_numbers)

# Preserved sorted list
save_sort = random_numbers[:]

# Apply the bubble sort algorithm
sorted_array, index_vector = bubble_sort(save_sort)
print("Sorted Array:", sorted_array)
print("Index Vector:", index_vector)

# Validation with Python's sort
sorted_by_python = sorted(random_numbers)
index_vector_python = [random_numbers.index(x) + 1 for x in sorted_by_python]
print("\nSorted Array Validation:", sorted_by_python)
print("Index Vector Validation:", index_vector_python)
```

```python
import pandas as pd
import numpy as np
from math import radians, cos, sin, sqrt, atan2
import matplotlib.pyplot as plt

# Load the store locations data
file_path = '/Users/dana/Desktop/Website/dflumer.github.io/Projects/ISC4221C/Lab_1/stores_location.dat'
stores_df = pd.read_csv(file_path, delim_whitespace=True)
stores_df.head()

def haversine(lat1, lon1, lat2, lon2):

    # Radius of Earth in kilometers
    R = 6371.0

    # Convert latitude and longitude from degrees to radians
    lat1_rad = radians(lat1)
    lon1_rad = radians(lon1)
    lat2_rad = radians(lat2)
    lon2_rad = radians(lon2)

    # Difference in coordinates
    dlat = lat2_rad - lat1_rad
    dlon = lon2_rad - lon1_rad

    # Haversine formula
    a = sin(dlat / 2)**2 + cos(lat1_rad) * cos(lat2_rad) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    # Distance in kilometers
    distance = R * c
    return distance

# Test with Tallahassee and Gainesville coordinates
tallahassee_coords = (30.4550, -84.2533)
gainesville_coords = (29.6520, -82.3250)

# Calculate distance and convert to miles
distance_km = haversine(*tallahassee_coords, *gainesville_coords)
distance_miles = distance_km * 0.621371  # 1 km = 0.621371 miles
distance_miles

# Coordinates for Ocala and Lake City Florida
ocala_coords = (29.1872, -82.1401)  # Ocala coordinates
lake_city_coords = (30.1897, -82.6393)  # Lake City coordinates

def calculate_distances(customer_coords, stores_df):
    distances = []
    for index, row in stores_df.iterrows():
        # Store coordinates
        store_coords = (row[2], -row[4])
        # Calculate distance
        distance = haversine(*customer_coords, *store_coords)
        distances.append((row[0], row[1], distance * 0.621371))  # Convert to miles
    # Sort by distance
    distances.sort(key=lambda x: x[2])
    return distances

# Calculate distances for Ocala and Lake City
distances_from_ocala = calculate_distances(ocala_coords, stores_df)
distances_from_lake_city = calculate_distances(lake_city_coords, stores_df)

distances_from_ocala, distances_from_lake_city
```
