```python

# 1. Random Number Generator

import numpy as np
import matplotlib.pyplot as plt

# Generate 10,000 uniform random coordinates
N = 10000
x = np.random.uniform(0, 1, N)
y = np.random.uniform(0, 1, N)

# Scatter plot
plt.scatter(x, y)
plt.title("Scatter")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

# Histogram
plt.hist(x, bins = 30, alpha = 0.5, label = "Value")
plt.hist(y, bins = 30, alpha = 0.5, label = "Count")
plt.title("Histogram")
plt.legend()
plt.show()

# Calculating means
x_mean = np.mean(x)
y_mean = np.mean(y)
print("The mean of x is", x_mean, "and the expected mean is 0.5.")
print("The mean of y is", y_mean, "and the expected mean is 0.5.")

# RMSE function
def calc_RMSE(points):
    x_mean, y_mean = np.mean(points, axis = 0)
    RMSE = np.sqrt((x_mean - 0.5)**2 + (y_mean - 0.5)**2) # RMSE formula
    return x_mean, y_mean, RMSE

# Calculate RMSE
RMSEs = [] # List to store RSME values
ns =range(10, 10001) # Range of N values

for n in ns:
    points = np.random.uniform(0, 1, (n, 2)) # Generate random points
    k, j, RMSE = calc_RMSE(points) # RMSE for each N
    RMSEs.append(RMSE)

# Plot RMSE on a log-log scale
plt.loglog(ns, RMSEs, label='RMSE')
plt.loglog(ns, [1/np.sqrt(n) for n in ns], label='N^-1/2') # Central Limit Theorem
plt.title("RMSE vs. N on a Log-Log Scale")
plt.xlabel("N")
plt.ylabel("RMSE")
plt.legend()
plt.show()
```

```python
# Monty Hall

def montyhall(trials):
    switch_wins = 0  # Count of wins when switching
    keep_wins = 0  # Count of wins when keeping original choice

    for i in range(trials):
        car_door = np.random.randint(1, 4)  # Random door with car
        # print(car_door)
        chosen_door = np.random.randint(1, 4)  # Contestant's chosen door
        # print(chosen_door)

        # Check if keeping or switching would win
        if car_door == chosen_door:
            keep_wins += 1
        else:
            switch_wins += 1

    return keep_wins, switch_wins

# Simulate 1000 trials
keep_wins, switch_wins = montyhall(1000)
print("Keeping wins:", keep_wins, "Switching wins:", switch_wins)
```
